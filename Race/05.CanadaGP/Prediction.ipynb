```python
import pandas as pd
import joblib
```

```python
CAN = pd.read_csv('Data/CanadaGP.csv')
```

```python
CAN
```

```python
CAN.drop(['Race_Result','Tyre_Compound','Sector1Time(s)','Sector2Time(s)','Sector3Time(s)','Longest_Stint'], axis=1,inplace=True)
```

```python
CAN
```

```python
CAN['FP2_BestTime(s)'].fillna(CAN['FP2_BestTime(s)'].median(),inplace=True)
CAN['Average_Laptime(s)'].fillna(CAN['Average_Laptime(s)'].mean(),inplace=True)

```

```python
CAN
```

```python
CAN.to_csv("Data/PredictionData.csv",index=False)
```

```python
model1 = joblib.load("Model/RandomForestModel92.pkl")
```

```python
X = CAN.iloc[:,1:]
```

```python
X
```

```python
y = model1.predict(X)
```

```python
df1 = CAN.copy()
```

```python
df1['Predicted_Standing'] = y
```

```python
df1
```

```python
df1 = df1.sort_values(by=['Predicted_Standing','Starting_Pos'])
df1 = df1.reset_index(drop=True)
df1['Predicted_Pos'] = df1.index+1
```

```python
df1
```

```python
df1.to_csv("Data/Predicted1.csv",index=False)
```
