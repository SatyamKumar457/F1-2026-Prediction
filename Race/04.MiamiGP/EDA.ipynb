```python
MIA  = pd.read_csv('Data/MiamiGP.csv')
```

```python
TRA = pd.concat([AUS,CHI,JAP],axis=0)
```

```python
TRA
```

```python
TRA.isnull().sum()
```

```python
TRA.info()
```

```python
TRA.dropna(subset=['Qualifying_Time(s)','Race_Result'],inplace=True)
```

```python
TRA.isnull().sum()
```

```python
TRA['FP2_BestTime(s)'].fillna(TRA['FP2_BestTime(s)'].median(),inplace=True)
TRA['Sector1Time(s)'].fillna(TRA['Sector1Time(s)'].median(),inplace=True)
TRA['Sector2Time(s)'].fillna(TRA['Sector2Time(s)'].median(),inplace=True)
TRA['Sector3Time(s)'].fillna(TRA['Sector3Time(s)'].median(),inplace=True)
TRA['Average_Laptime(s)'].fillna(TRA['Average_Laptime(s)'].mean(),inplace=True)
TRA['Longest_Stint'].fillna(TRA['Longest_Stint'].median(),inplace=True)
TRA['Tyre_Compound'].fillna(TRA['Tyre_Compound'].mode()[0],inplace=True)
```

```python
TRA.isnull().sum()
```

```python
TRA.columns
```

```python
PRE = MIA.copy()
```

```python
PRE.isnull().sum()
```

```python
PRE['FP2_BestTime(s)'].fillna(PRE['FP2_BestTime(s)'].median(),inplace=True)
PRE['FP3_BestTime(s)'].fillna(PRE['FP3_BestTime(s)'].median(),inplace=True)
```

```python
PRE.isnull().sum()
```

```python
PRE
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

encoder.fit(TRA[['Tyre_Compound']])

TRA_encoded = encoder.transform(TRA[['Tyre_Compound']])
PRE_encoded = encoder.transform(PRE[['Tyre_Compound']])
```

```python
cols = encoder.get_feature_names_out(['Tyre_Compound'])

TRA_encoded = pd.DataFrame(TRA_encoded, columns=cols, index=TRA.index)
PRE_encoded = pd.DataFrame(PRE_encoded, columns=cols, index=PRE.index)
```

```python

TRA = TRA.drop('Tyre_Compound', axis=1).join(TRA_encoded)
PRE = PRE.drop('Tyre_Compound', axis=1).join(PRE_encoded)
```

```python
TRA
```

```python
PRE
```

```python
TRA.to_csv('TrainingData.csv',index=False)
PRE.to_csv('PredictionData.csv',index=False)
```
