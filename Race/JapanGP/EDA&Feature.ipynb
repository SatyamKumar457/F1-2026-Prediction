```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

# Australia GP (For Training)

```python
Aus = pd.read_csv('Data/AustraliaGP.csv')
```

```python
Aus
```

```python
Aus.info()
```

```python
imputer = SimpleImputer(strategy='mean')
Aus['FP2_BestTime(s)'] = imputer.fit_transform(Aus[['FP2_BestTime(s)']])
Aus['FP3_BestTime(s)'] = imputer.fit_transform(Aus[['FP3_BestTime(s)']])
```

```python
Aus
```

```python
Aus.info()
```

```python
Aus['Qualifying_Time(s)'].max()
```

```python
Aus['Qualifying_Time(s)'].fillna(Aus['Qualifying_Time(s)'].max(),inplace=True)
```

```python
Aus
```

```python
Aus['Starting_Pos'].fillna(20,inplace=True)
```

```python
Aus
```

```python
Aus = pd.get_dummies(Aus, columns=["Tyre_Compound"], drop_first=True)
```

```python
Aus
```

```python
Aus.to_csv('AustraliaGPModified.csv',index=False)
```

# Japan GP (For Prediction)

```python
Jap = pd.read_csv('Data/JapanGP.csv')
```

```python
Jap
```

```python
Jap = Jap[Jap["Starting_Pos"].notna()]
```

```python
Jap
```

```python
Jap['FP2_BestTime(s)'] = imputer.fit_transform(Jap[['FP2_BestTime(s)']])
Jap['Sector1Time(s)'] = imputer.fit_transform(Jap[['Sector1Time(s)']])
Jap['Average_Laptime(s)'] = imputer.fit_transform(Jap[['Average_Laptime(s)']])
```
